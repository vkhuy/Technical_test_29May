```python
import matplotlib.pyplot as plt
seq = "CCCTACTTCACTCCAGTACTATAGTAGTAGCTGGGGTTTTCTTACTTATTCGCTTCCACCCACCAATAGAAAATAATACAACAATTCAAAGTCTTACACTATGCCTAGGAGCTATTACTACCATATTCATAGCAATCTGCGCCCTAACACAAAATGACATTAAAAAAATTGTAGCCTTCTCTACCTCAA"
C_count = seq.count("C")
A_count = seq.count("A")
T_count = seq.count("T")
G_count = seq.count("G")
print("C = " + str(C_count))
print("A = " + str(A_count))
print("T = " + str(T_count))
print("G = " + str(G_count))



base_count = [C_count,A_count,T_count,G_count]
base=["C","A","T","G"]
positions = range(len(base))

plt.bar(positions,base_count)
plt.xticks (positions,base)
plt.xlabel("Bases")
plt.ylabel("Occurrences")
plt.show()
```
